```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nodule_counts = pd.read_csv('data/lidc-idri-nodule-counts.csv', usecols=[0,1,2,3])
list32 = pd.read_csv('data/list3.2.csv', usecols=[0,1,2,3,4,5,6,7])
diag_data = pd.read_csv('data/tcia-diagnosis-data.csv')
```

```python
# remove bad row
nodule_counts = nodule_counts[nodule_counts['TCIA Patent ID'].isnull()==False]
```

```python
nodule_counts.head()
```

```python
nodule_counts.sum() 
```

```python
fig, ax = plt.subplots(1,3, figsize = (16,4))
ax[0].hist(nodule_counts['Total Number of Nodules* '], bins=20);
ax[1].hist(nodule_counts['Number of Nodules <3mm'], bins=20);
ax[2].hist(nodule_counts['Number of Nodules >=3mm'], bins=20);
ax[0].grid()
ax[1].grid()
ax[2].grid()
ax[0].set_title('Histogram of all nodules')
ax[1].set_title('Histogram of all nodules < 3mm')
ax[2].set_title('Histogram of all nodules > 3mm')
```
